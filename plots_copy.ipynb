# top 50 most popular songs

```python
top50 = data.sort_values(by = 'popularity', ascending = False).iloc[0:50]
top50.head()
```

```python
from urllib.parse import urlencode

def make_youtube_query(name):
    return "https://www.youtube.com/search?" + urlencode({'q': '"{0}"'.format(name)})


top50['url'] = top50['track_name'].apply(make_youtube_query)

top50bar = alt.Chart(top50).mark_bar().encode(
    alt.X("popularity:Q",scale=alt.Scale(domain=[85, 100])),
    alt.Y("track_name:N", sort = '-x'),
    tooltip=['track_name:N', "popularity:Q", 'url:N'],
    # href = 'url:N'
)

top50bar
```

```python
# Create a sample dataframe
top50bar = alt.Chart(top50).mark_bar().encode(
    alt.Y("popularity:Q",scale=alt.Scale(domain=[85, 100])),
    alt.X("track_name:N", sort = '-y'),
    tooltip=['track_name:N', "popularity:Q", 'url:N'],
    # href = alt.Href('url:N'),
)

top50bar
```
