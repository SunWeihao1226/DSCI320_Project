## Exploratory Data Analysis


```python
import pandas as pd
import altair as alt
```

## Read the data

```python
url = 'https://raw.githubusercontent.com/luckyberen/DSCI320_Project/main/spotify_tracks.csv'
data = pd.read_csv(url)
data.head()
```

## Summary of each variable

```python
print(data.info())
```

```python
data.describe()
```

## Visualizations of variables

### Nominal variables

```python
alt.data_transformers.disable_max_rows()


plot_nominal = alt.Chart(data).mark_bar().encode(
    alt.X(alt.repeat('repeat'), type='nominal'),
    alt.Y('count()')
).properties(
  width=200,
  height=200
).repeat(
  data=data,
  repeat = ['explicit','mode','key', 'time_signature'],
  columns = 2
)


plot_nominal
```

## Quantitative variables

```python
plot_quantitative = alt.Chart(data).mark_bar().encode(
    alt.X(alt.repeat('repeat'), type='quantitative',bin=alt.BinParams(maxbins=30)),
    alt.Y('count()')
).properties(
  width=200,
  height=200
).repeat(
  data=data,
  repeat = ['tempo','duartion_s','popularity','danceability','loudness','speechiness','acousticness',
            'instrumentalness','liveness','energy','valence'],
  columns = 3
)
plot_quantitative
```
